<center> <h1><a class="anchor" id="part2-2"> PARTE 2: TESTE DE HIPOTESE</a></h1> </center>

```python
# Função para calcular a diferença de médias
def diff_means(sample1, sample2):
    return np.mean(sample1) - np.mean(sample2)

# Função para realizar bootstrap e calcular intervalo de confiança
def bootstrap_CI_diff_means(sample1, sample2, num_samples, alpha):
    diff_means_samples = []
    combined_sample = np.concatenate([sample1, sample2])
    n1 = len(sample1)
    
    for _ in range(num_samples):
        bootstrap_sample1 = np.random.choice(combined_sample, size=n1, replace=True)
        bootstrap_sample2 = np.random.choice(combined_sample, size=n1, replace=True)
        diff_means_samples.append(diff_means(bootstrap_sample1, bootstrap_sample2))
    
    # Cálculo do intervalo de confiança
    lower_percentile = (alpha / 2) * 100
    upper_percentile = 100 - lower_percentile
    lower_bound = np.percentile(diff_means_samples, lower_percentile)
    upper_bound = np.percentile(diff_means_samples, upper_percentile)
    
    return diff_means_samples, lower_bound, upper_bound

# Separar os preços das casas térreas e das casas com mais de um andar
preco_terrea = df[df['floors'] == 1]['price']
preco_mais_de_um_andar = df[df['floors'] > 1]['price']

# Calcular a diferença de médias observada
observed_diff_means = diff_means(preco_terrea, preco_mais_de_um_andar)

# Realizar bootstrap e calcular intervalo de confiança
num_samples = 1000  # Número de amostras bootstrap
alpha = 0.05  # Nível de significância
diff_means_samples, lower_bound, upper_bound = bootstrap_CI_diff_means(preco_terrea, preco_mais_de_um_andar, num_samples, alpha)

# Calcular p-valor da área abaixo do valor observado
p_value = np.mean(np.array(diff_means_samples) <= observed_diff_means)

# Verificar se o p-valor é menor que 5%
if p_value < 0.05:
    print("O p-valor ({:.2f}) é menor que 5%, portanto, rejeitamos a hipótese nula.".format(p_value))
else:
    print("O p-valor ({:.2f}) não é menor que 5%, portanto, não rejeitamos a hipótese nula.".format(p_value))

# Plotar histograma com intervalo de confiança e valor observado
plt.hist(diff_means_samples, bins=20)
plt.axvline(x=upper_bound, color='gold', linestyle='--', label='Limite superior do intervalo de confiança (95%)')
plt.axvline(x=lower_bound, color='gold', linestyle='--', label='Limite inferior do intervalo de confiança (95%)')
plt.axvline(x=observed_diff_means, color='red', linestyle='-', label='Valor observado')
plt.title('Distribuição das diferenças de médias')
plt.xlabel('Diferença de médias')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
# Definir o número de folds para a validação cruzada (vamos usar 5 folds)
num_folds = 5

# Criar os objetos KFold para controlar a validação cruzada
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Realizar a validação cruzada e calcular a métrica desejada (por exemplo, R^2)
scores = cross_val_score(xgbr, x, y, cv=kf, scoring='r2')

# Exibir os resultados da validação cruzada
print(f'Média R^2: {np.mean(scores)}')
print(f'Desvio padrão R^2: {np.std(scores)}')
```
